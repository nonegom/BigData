 ## 실습1. 데이터 선택
1. 데이터 구성하기
2. slice: B열선택 / B,D열 선택 / 4~8행 선택
3. loc: 2,8행 선택 A,C열 선택 / A,C열 선택 / 3~8행 선택 / 3행 선택(단일행)
   - iloc: 0,6행, A열 선택 / A열 선택 / 2~6행 선택 / 5행 선택(단일행)
4. query:  D > 0.5 / B > D _slice 표현으로 / A > 0.5 and D<0.0 / A > 0.5 or D<0.0

```python
# 1. 데이터 구성하기
import numpy as np
import pandas as pd
from numpy.random import randn

np.random.seed(1234)
df = pd.DataFrame(randn(10, 4), columns=['A','B','C','D'])
df.head()
```

```python
# 2. slice: B열선택 / B,D열 선택 / 4~8행 선택
df[['B']]
df[['B','D']]
df[4:9]
```

```python
# 3. loc: 2,8행 선택 A,C열 선택 / A,C열 선택 / 3~8행 선택 / 3행 선택(단일행)

df.loc[2:8, ["A","C"]]
df.loc[:,["A","C"]]
df.loc[3:8,:]
df.loc[3:3]

# 3.  iloc: 0,6행, A열 선택 / A열 선택 / 2~6행 선택 / 5행 선택(단일행)
df.iloc[0:7, [0]]
df.iloc[:,[0]]
df.iloc[2:7]
df.iloc[5:6]
```

```python
#4. query:  D > 0.5 / B > D _slice 표현으로 / A > 0.5 and D<0.0 / A > 0.5 or D<0.0

df.query('D > 0.5')
df[df.B > df.D]
df.query('A > 0.5 and D < 0.0') # == df.query('A > 0.5 & D < 0.0')
df.query('A > 0.5 or D < 0.0')  # == df.query('A > 0.5 | D < 0.0')
```

 ## 실습2. 데이터 결합
 1. 데이터 구성하기
 2. concat을 이용한 데이터 결합
 3. parameter 변경

```python
# 1. 데이터 구성하기
df1 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df2 = pd.DataFrame({
        "E": ["E4", "E5", "E6", "E7"],
        "F": ["F4", "F5", "F6", "F7"],
        "G": ["G4", "G5", "G6", "G7"],
        "H": ["H4", "H5", "H6", "H7"]},
        index=[0, 1, 2, 3])

df3 = pd.DataFrame({
        "I": ["I8", "I9", "I10", "I11"],
        "J": ["J8", "J9", "J10", "J11"],
        "K": ["K8", "K9", "K10", "K11"],
        "L": ["L8", "L9", "L10", "L11"]},
        index=[0, 1, 2, 3])
```

```python
# 2. concat을 이용한 데이터 결함
result = pd.concat([df1, df2, df3], axis = 1)
result.head()
```

```python
# 3. parameter 변경
column_name = [i + "열" for i in result.columns]
result.columns = column_name
result.head()
```

 ## 실습4. 데이터 변환
 1. 데이터 구성하기
 2. Groupby, agg를 이용한 평균값 계산
 3. pivot을 이용한 변환

```python
# 1. 데이터 구성하기 (원본 데이터가 책과 달라진 관계로 column 변경)

df = pd.read_csv("./Data/FITNESS.csv", engine = "python")
df.head()

df["GENDER"] =  ["남성" if x == "M" else "여성" for x in df["GENDER"] ]
df["AGEGROUP"] = [str(i)+"대" for i in df["AGEGROUP"]]
df.head()
```

```python
weight_mean = df.groupby(['GENDER', 'AGEGROUP'])["WEIGHT"].agg(["mean"]).unstack()
```

```python
# groupby & aggfunc이용
weight_mean
```

```python
# pivot 이용
df.pivot_table(index = 'GENDER', columns="AGEGROUP", values= 'WEIGHT', aggfunc="mean")
```
