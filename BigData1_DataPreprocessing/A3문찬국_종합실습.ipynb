```python
import pandas as pd
import numpy as np

pop_seoul = pd.read_csv('./Data/서울인구통계.txt', sep="\t", engine='python',
                       thousands=',', encoding="utf-8")
pop_seoul.head()
```

```python
# drop?
pop1 = pop_seoul[['자치구', '인구','인구.3', '인구.6', '65세이상고령자']]
pop_seoul = pop1.drop([0,1])
pop_seoul.head()
```

```python
# column변경
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                          pop_seoul.columns[1]:'인구수',
                          pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자',}, inplace=True)
pop_seoul.head()
```

```python
cctv_seoul = pd.read_excel('./Data/서울시CCTV.xlsx')
cctv_seoul.head()
```

```python
cctv_seoul.columns
```

```python
# 열 이름 변경
cctv_seoul.rename(columns={cctv_seoul.columns[0]:'구별'}, inplace = True)
cctv_seoul.rename(columns={cctv_seoul.columns[1]:'cctv수'}, inplace = True)

cctv_seoul.head()
```

```python
# 데이터 정렬
cctv_seoul.sort_values(by='cctv수', ascending=True).head(10) # 오름차순
cctv_seoul.sort_values(by='cctv수', ascending=False).head(10)# 내림차순
```

```python
# NaN값 확인
cctv_seoul.isnull().sum() # NaN 값이 있는 개수 확인
cctv_seoul.loc[cctv_seoul.isnull()['2011년 이전']]
```

```python
# 새로운 칼럼 생성
cctv_seoul['최근증가율']= (cctv_seoul['2018년'] + cctv_seoul['2017년']  + cctv_seoul['2016년']) /(cctv_seoul['cctv수']-(cctv_seoul['2018년'] + cctv_seoul['2017년']  + cctv_seoul['2016년']))*100
```

```python
cctv_seoul.sort_values(by="최근증가율", ascending=False).head(5)
```

## 서울시 인구 데이터 파악

```python
pop_seoul.head()
```

```python
# index 2 :[합계]행 삭제
pop_seoul.drop([2], inplace = True)
pop_seoul.head()
```

```python
pop_seoul.dtypes # object 타입 변경
pop_seoul = pop_seoul.astype({"인구수": "int", "한국인": "int", 
                              "외국인": "int", "고령자": "int"})
```

```python
# 외국인 비율과 고령자 비율 추가
pop_seoul['외국인 비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자 비율'] = pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by='외국인 비율', ascending=False).head()
```

```python
pop_seoul.sort_values(by='고령자 비율', ascending=False).head()
```

```python
ls = sorted(list(pop_seoul.구별.unique()))
```

```python
cctv_seoul['구별'] = ls
cctv_seoul.head()
```

### 데이터 합치기

```python
result = pd.merge(cctv_seoul, pop_seoul, on ='구별')
result.head()
```

```python
del result['2011년 이전']
```

```python
# drop으로 열도 제거할 수 있다.
result.drop(['2012년', '2013년', '2014년', '2015년',
             '2016년', '2017년', '2018년'], axis=1, inplace = True)
```
