```python
ds_Exh_QC1 = pd.read_csv("./Data/EXH_QC1.csv", engine = 'python')
ds_Exh_QC1.head()
```

```python
df = ds_Exh_QC1[['Flaws', 'Period']]
df.head()
```

```python
count = df['Flaws'].value_counts().sort_index()
count
```

```python
percent = count / sum(count)* 100
percent
```

```python
cumcnt = np.cumsum(count)
cumcnt
```

```python
cumpct = np.cumsum(percent)
cumpct
```

```python
count_data = pd.DataFrame({'Count': count, 'Percent': percent, 
                           'Cumcnt': cumcnt, 'CumPct': cumpct})
count_data.columns.name = 'Flaws'
count_data
```

## 실습

```python
df = ds_mycars[['model', 'mpg']]
count = df['model'].value_counts().sort_index()
cumcnt = np.cumsum(count)
percent = count / sum(count) * 100
cumpct = np.cumsum(percent)

df = pd.DataFrame({'Count': count, 'Percent': percent, 
                    'Cumcnt': cumcnt, 'CumPct': cumpct})
df.columns.name = 'mpodel'
```

```python
df = ds_mycars[['driving', 'mpg']]
count = df['driving'].value_counts().sort_index()
cumcnt = np.cumsum(count)
percent = count / sum(count) * 100
cumpct = np.cumsum(percent)

df = pd.DataFrame({'Count': count, 'Percent': percent, 
                    'Cumcnt': cumcnt, 'CumPct': cumpct})
df.columns.name = 'driving'
```
